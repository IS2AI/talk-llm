```python
from IPython.display import display as dp
from IPython.display import Audio
import subprocess
import random
import os 
```

```python
libritts_model = "piper_tts_models/libritts/en_US-libritts-high.onnx"
libritts_config = "piper_tts_models/libritts/en_US-libritts-high.onnx.json"

libritts_r_model = "piper_tts_models/libritts-r/en_US-libritts_r-medium.onnx"
libritts_r_config = "piper_tts_models/libritts-r/en_US-libritts_r-medium.onnx.json"

vctk_model = "piper_tts_models/vctk/en_GB-vctk-medium.onnx"
vctk_config = "piper_tts_models/vctk/en_GB-vctk-medium.onnx.json"
```

```python
target_word = "issai"
other_words = "other"

num_libritts_speakers = 904
num_vctk_speakers = 109

num_random_words = 10

save_path = "synthetic_tts_dataset"

for label in [target_word, other_words]:
    save_label_path = os.path.join(save_path, label)
    if not os.path.exists(save_label_path):
        os.makedirs(save_label_path)
```

```python
# 1000 most common English words
english_words_file = open("1-1000.txt", "r") 
english_words_list = english_words_file.read() 
english_words_list = english_words_list.split("\n") 
english_words_file.close() 

print("Number of words:", len(english_words_list))
```

```python
for i in range(num_libritts_speakers):
    print("Speaker: {}/{}".format(i+1,num_libritts_speakers))
    wav_file = os.path.join(save_path, target_word, "libritts_{}.wav".format(i))
    command = f'echo "{target_word}" | ./piper_lib/piper --model {libritts_model} --config {libritts_config} --speaker {i} --output_file {wav_file}'
    subprocess.run(command, shell=True, check=True)
    
    wav_file = os.path.join(save_path, target_word, "librittsr_{}.wav".format(i))
    command = f'echo "{target_word}" | ./piper_lib/piper --model {libritts_r_model} --config {libritts_r_config} --speaker {i} --output_file {wav_file}'
    subprocess.run(command, shell=True, check=True)
```

```python
for i in range(num_vctk_speakers):
    print("Speaker: {}/{}".format(i+1,num_vctk_speakers))
    wav_file = os.path.join(save_path, target_word, "vctk_{}.wav".format(i))
    command = f'echo "{target_word}" | ./piper_lib/piper --model {vctk_model} --config {vctk_config} --speaker {i} --output_file {wav_file}'
    subprocess.run(command, shell=True, check=True)
```

```python
for i in range(num_libritts_speakers):
    print("Libritts Speaker: {}/{}".format(i+1,num_libritts_speakers))
    for j in range(num_random_words):
        rand_int = random.randint(0, len(english_words_list)-1)
        rand_word = english_words_list[rand_int]
        wav_file = os.path.join(save_path, other_words, "libritts_{}_{}.wav".format(i,j))
        command = f'echo "{rand_word}" | ./piper_lib/piper --model {libritts_model} --config {libritts_config} --speaker {i} --output_file {wav_file}'
        subprocess.run(command, shell=True, check=True)
    
    print("Libritts-r Speaker: {}/{}".format(i+1,num_libritts_speakers))
    for j in range(num_random_words):
        rand_int = random.randint(0, len(english_words_list)-1)
        rand_word = english_words_list[rand_int]
        wav_file = os.path.join(save_path, other_words, "librittsr_{}_{}.wav".format(i,j))
        command = f'echo "{rand_word}" | ./piper_lib/piper --model {libritts_r_model} --config {libritts_r_config} --speaker {i} --output_file {wav_file}'
        subprocess.run(command, shell=True, check=True)
```

```python
for i in range(num_vctk_speakers):
    print("VCTK Speaker: {}/{}".format(i+1,num_vctk_speakers))
    for j in range(num_random_words):
        rand_int = random.randint(0, len(english_words_list)-1)
        rand_word = english_words_list[rand_int]
        wav_file = os.path.join(save_path, other_words, "vctk_{}_{}.wav".format(i,j))
        command = f'echo "{rand_word}" | ./piper_lib/piper --model {vctk_model} --config {vctk_config} --speaker {i} --output_file {wav_file}'
        subprocess.run(command, shell=True, check=True)
```
